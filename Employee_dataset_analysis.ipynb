Data Filtering and Selection

```python
# Extract all rows where the department is "Finance."

ds[ds["Department"]=="Finance"]
```

```python
# Find how many employees are there in each country.

ds["Country"].value_counts()
```

```python
# List all employees earning an annual salary greater than $100,000.

ds["Annual Salary"]=ds["Annual Salary"].str.replace("[$,]","",regex=True).astype(int)
ds[ds["Annual Salary"]>100000]
```

```python
# Find all employees with the job title "Data Scientist."

ds[ds["Job Title"]=="Data Scientist"]
```

Aggreration and Grouping

```python
# Calculate the average annual salary for each department.

ds.groupby("Department")["Annual Salary"].mean()
```

```python
# Find the average salary and bonus percentage for each gender.

ds["Annual Salary"] = ds["Annual Salary"].astype(str)
ds["Annual Salary"]=ds["Annual Salary"].str.replace("[$,]","",regex=True).astype(int)

ds["Bonus %"]=ds["Bonus %"].astype(str)
ds["Bonus %"]=ds["Bonus %"].str.replace("[%]","",regex=True).astype(int)

ds.groupby("Gender")[["Annual Salary","Bonus %"]].mean()
```

```python
# Determine the average age of employees in each business unit.

ds.groupby("Business Unit")["Age"].mean()
```

Data Transformation

```python
# Create two new columns for first and last names by splitting the full name column.

# df[['first name', 'last name']] = df['full name'].str.split(' ', 1, expand=True)

ds["First Name"]=ds["Full Name"].str.split(" ").str[0]
ds["Last Name"]=ds["Full Name"].str.split(" ").str[1]
ds
```

```python
# Create a new column that calculates the number of years an employee has worked (from hire date to exit date or today if still employed).

# Convert object to datetime.
from datetime import datetime

ds["Exit Date"]=pd.to_datetime(ds["Exit Date"], format="%m/%d/%Y", errors="coerce")
ds["Hire Date"]=pd.to_datetime(ds["Hire Date"], format="%m/%d/%Y", errors="coerce")

# Replace null ny today's date.
ds["Exit Date"]=ds["Exit Date"].fillna(pd.Timestamp.today().normalize())

ds["Years Worked"]=ds["Exit Date"].dt.year-ds["Hire Date"].dt.year
print(ds["Years Worked"])
```

Data Visualization

```python
# Plot a histogram to show the distribution of annual salaries.

ds["Annual Salary"] = ds["Annual Salary"].astype(str)
ds["Annual Salary"]=ds["Annual Salary"].str.replace("[$,]","",regex=True).astype(int)

ds["Annual Salary"].plot(kind="hist", bins=30, title="Salary Distribution")
plt.show()
# bins means the number of intervals the data is divided into.
```

```python
# Create a bar chart to compare average salaries across departments.

ds.groupby("Department")["Annual Salary"].mean().plot(kind="bar",title="Average Salary by Department")
plt.show()
```

```python
# Create a pie chart to display the gender proportion in the dataset.

ds["Gender"].value_counts().plot(kind="pie",title="Gender Proportion", autopct="%1.1f%%")
plt.show()

# autopct is used as 1.1=51.8 and 5.5=51.80000.
```

Advanced Filtering

```python
# Identify the top 5 employees with the longest service duration.

ds.sort_values("Years Worked",ascending=False).head(5)
```

```python
# Find employees with a bonus percentage greater than 10% and an annual salary below $80,000.

ds["Bonus %"]=ds["Bonus %"].astype(str)
ds["Bonus %"]=ds["Bonus %"].str.replace("[%]","",regex=True).astype(int)

ds[(ds["Bonus %"]>10) & (ds["Annual Salary"]>80000)]
```

Advanced Questions

```python
# Count the number of unique ethnicities represented in each department.

# This gives number of people in each department.
ds["Department"].value_counts()

# The answer of question above:
ds.groupby("Department")["Ethnicity"].nunique()
```

```python
# Find the top 3 departments where the total compensation (annual salary + bonus) is the highest, and display their total compensation.

ds["Total Compensation"]=ds["Annual Salary"]+ds["Bonus %"]
ds.groupby("Department")["Total Compensation"].sum().sort_values(ascending=False).head(3)
```

```python
# Calculate the retention rate for each combination of gender and ethnicity. Retention rate is the percentage of employees still working (i.e., exit date is missing).

# Number of employees still working:
# ds["Exit Date"].isna().sum()

retention=ds["Exit Date"].isna()
retention_rate=ds.groupby(["Gender","Ethnicity"]).apply(lambda x: x.isna().mean() * 100)
print(retention_rate)
```

```python
# For each department, identify the most common job title.

ds.groupby("Department")["Job Title"].value_counts()

# Or, we can do:

ds.groupby("Department")["Job Title"].apply(lambda x: x.value_counts().idxmax())
```

```python
# Calculate the average time (in years) employees worked before exiting for each department.

ds.groupby("Department")["Years Worked"].mean()
```

```python
# Compare the average salary for each job title between genders.

ds.groupby(["Gender","Job Title"])["Annual Salary"].mean()
```
